```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
print("To understand,, how the pair plot works, lets import the built-in data from seaborn library.")
data = sns.load_dataset("penguins")
data.head()
```

```python
sns.pairplot(data, height=3)
plt.show()
```

Notice pair plot automatically plots all the numeric columns only. Also, the default value of kind parameter in pairplot is scatter plot.

Note that we can also use hue variable as input.

```python
sns.pairplot(data, hue="species", height=3)
plt.show()
```

```python
print("If we want to use histogram instead of kde in above pairplot diagonally, we can use diag_kind='hist' as input")
```

```python
sns.pairplot(data, hue="species", height=3, diag_kind="hist")
plt.show()
```

```python
print("Furthermore, we can use kind='kde' as input for kde pairplot. Also, we can use kind='hist' for histogram pairplot. \nIf want histogram diagonally, we \ncan use diag_kind='hist' as input.")
sns.pairplot(data, height=3, kind='kde')
plt.show()
```

If we want to graph pairplot only for selected number of variables, we can use x_vars and y_vars paramters to specify variables along x-axis and y-axis.

```python
sns.pairplot(data, x_vars=["bill_length_mm", "bill_depth_mm", "flipper_length_mm"], 
             y_vars=["bill_length_mm", "bill_depth_mm"], height=2)
plt.show()
```

Note that to cutomize the off-diagonal and diagonal plots of pairplot, we can use plot_kws and diag_kws parameters that accept dicts of keyword arguments to customize the off-diagonal and diagonal plots, respectively.

```python
sns.pairplot(data, hue="species", height=3, diag_kind="hist", plot_kws=dict(size=data["island"],marker="o", alpha=0.7,linewidth=1), diag_kws=dict(element="step"))
plt.show()
```
